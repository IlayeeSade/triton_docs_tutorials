```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV data into a pandas DataFrame
df = pd.read_csv('rbf-performance.csv')

# Create the plot
plt.figure(figsize=(12, 8))

# Plot each method with different colors and labels
plt.plot(df['M'], df['Regular'], color='orange', label='Regular', marker='o')
plt.plot(df['M'], df['PyTorch'], color='green', label='PyTorch', marker='o')
plt.plot(df['M'], df['Triton2'], color='red', label='Triton2', marker='o')

# Customize the plot
plt.xlabel('Matrix Size (M = N = K)')
plt.ylabel('Execution Time (seconds)')
plt.title('Performance Comparison of Different Methods')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Use logarithmic scale for y-axis due to large differences in values
# plt.yscale('log')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Save the plot
plt.savefig('matrix_performance_comparison.png')
plt.show()
```
